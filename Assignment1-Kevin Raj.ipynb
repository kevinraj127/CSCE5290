# Code without stop words

```python
from bs4 import BeautifulSoup
import urllib.request
import nltk
from nltk.corpus import stopwords

#use Wikipedia SpaceX page for processing
response = urllib.request.urlopen('https://en.wikipedia.org/wiki/SpaceX')
html = response.read()
soup = BeautifulSoup(html,"html5lib")
text = soup.get_text(strip=True)

#Split text into separate tokens
tokens = [t for t in text.split()]
clean_tokens = tokens[:]
sr = stopwords.words('english')

#Remove stop words
for token in tokens:
    if token in stopwords.words('english'):
        clean_tokens.remove(token)
freq = nltk.FreqDist(clean_tokens)

#Remove word occurrences that are less than 5
for key,val in list(freq.items()):
    if val < 5: 
        del freq[key]
for key, val in freq.items():
    print(str(key) + ':' + str(val))
```

# Frequency plot without stop words

```python
freq.plot(10, cumulative=False)
```

# Another visualization for NLP code without stop words

```python
freq.tabulate(10, cumulative=False)
```
